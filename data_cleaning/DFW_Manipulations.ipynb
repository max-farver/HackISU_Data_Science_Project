```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('DFW_fully_cleaned.csv')
df.head()
df.shape
```

```python
temprating = ""
i=0
diffs = []
for index, row in df.iterrows():
    if int(row['% DFW']) <= 7:
        temprating = '1'
    elif int(row['% DFW']) <= 11:
        temprating = '2'
    elif int(row['% DFW']) <= 15:
        temprating = '3'
    elif int(row['% DFW']) <= 19:
        temprating = '4'
    elif int(row['% DFW']) <= 23:
        temprating = '5'
    elif int(row['% DFW']) <= 27:
        temprating = '6'
    elif int(row['% DFW']) <= 33:
        temprating = '7'
    elif int(row['% DFW']) <= 35:
        temprating = '8'
    elif int(row['% DFW']) <= 39:
        temprating = '9'
    elif int(row['% DFW']) >= 40:
        temprating = '10'
    str(temprating)
    print(temprating)
    diffs.append(temprating)
    i+=1
```

```python
i = 0
for x in diffs:
    df.at[i,'Difficulty'] = x
    i+=1
```

```python
df.to_csv("DFW_with_rating.csv")
```
