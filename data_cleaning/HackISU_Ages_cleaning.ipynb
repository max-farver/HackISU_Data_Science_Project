```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("all-ages.csv")
df.head()
```

```python
df.boxplot(by='Major_category', column='Median', figsize=(12,8), rot=90)
```

```python
df.loc[df['Major_category'] == 'Interdisciplinary']
```

```python
#remove pointless majors
df = df[df['Major_category'] != 'Interdisciplinary']
```

```python
plt = df.boxplot(by='Major_category', column='Median', figsize=(12,8), rot=90)
plt.set_title('Field vs Income')
plt.set_ylabel('Income')
```

```python
df.boxplot(by='Major_category', column='Employed', figsize=(12,8), rot=90)
```
